```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


#import OpenCV module
import cv2
#import os module for reading training data directories and paths
import os
#import numpy to convert python lists to numpy arrays as 
#it is needed by OpenCV face recognizers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
get_ipython().run_line_magic('matplotlib', 'inline')

import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

from PIL import Image
from numpy import* 
import tensorflow as tf


from skimage import color
from skimage import io
import cv2
import numpy as np
from matplotlib import pyplot as plt


# In[2]:


def prepare_training_data(data_folder_path):
    
    #------STEP-1--------
    #get the directories (one directory for each subject) in data folder
    dirs = os.listdir(data_folder_path)
    
    #list to hold all subject faces
    faces = []
    #list to hold labels for all subjects
    labels = []
    
    #let's go through each directory and read images within it
    for dir_name in dirs:
        
       
            
        #------STEP-2--------
        #extract label number of subject from dir_name
        #format of dir name = slabel
        #, so removing letter 's' from dir_name will give us label
        label = dir_name
        
        #build path of directory containin images for current subject subject
        #sample subject_dir_path = "training-data/s1"
        subject_dir_path = data_folder_path + "/" + dir_name
        
        #get the images names that are inside the given subject directory
        subject_images_names = os.listdir(subject_dir_path)
        
        #------STEP-3--------
        #go through each image name, read image, 
        #detect face and add face to list of faces
        for image_name in subject_images_names:
            
            
            
            #build image path
            #sample image path = training-data/s1/1.pgm
            image_path = subject_dir_path + "/" + image_name

            #read image
            #temp = cv2.imread(image_path)
           # temp = Image.open(image_path)
            img = io.imread(image_path)
            imgGray = color.rgb2gray(img)
            #temp=Image.open('a.jpg') 
           

                     
            #------STEP-4--------
            #for the purpose of this tutorial
            #we will ignore faces that are not detected
            if imgGray is not None:
                #add face to list of faces
                
                faces.append(imgGray)
               
                #add label for this face
                labels.append(label)
              
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    
    print(labels)

    return faces, labels
            
           


# In[3]:


print("Preparing data...")
faces, labels = prepare_training_data("C://Users//Amrutha Y//Desktop//Medicinal Leaf dataset")
print("Data prepared")


# In[4]:


faces1=np.array(faces)
labels1=np.array(labels)
print(faces1.ndim)
print("pics: ", faces1.shape)
print("labels: ", labels1.shape)


# In[5]:


Xdata = faces1 # store images in Xdata
Ydata = labels1.reshape(-1,1) # store labels in Ydata


# In[6]:


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Xdata, Ydata, test_size = 0.2, random_state=46)

print("x_train: ",x_train.shape)
print("x_test: ",x_test.shape)
print("y_train: ",y_train.shape)
print("y_test: ",y_test.shape)


# In[7]:


import cv2
import numpy as np
from matplotlib import pyplot as plt

	
def get_pixel(x_train, center, x, y):
	
	new_value = 0
	
	try:
		# If local neighbourhood pixel
		# value is greater than or equal
		# to center pixel values then
		# set it to 1
		if x_train[x][y] >= center:
			new_value = 1
			
	except:
		# Exception is required when
		# neighbourhood value of a center
		# pixel value is null i.e. values
		# present at boundaries.
		pass
	
	return new_value

# Function for calculating LBP
def lbp_calculated_pixel(x_train, x, y):

	center = x_train[x][y]

	val_ar = []
	
	# top_left
	val_ar.append(get_pixel(x_train, center, x-1, y-1))
	
	# top
	val_ar.append(get_pixel(x_train, center, x-1, y))
	
	# top_right
	val_ar.append(get_pixel(x_train, center, x-1, y + 1))
	
	# right
	val_ar.append(get_pixel(x_train, center, x, y + 1))
	
	# bottom_right
	val_ar.append(get_pixel(x_train, center, x + 1, y + 1))
	
	# bottom
	val_ar.append(get_pixel(x_train, center, x + 1, y))
	
	# bottom_left
	val_ar.append(get_pixel(x_train, center, x + 1, y-1))
	
	# left
	val_ar.append(get_pixel(x_train, center, x, y-1))
	
	# Now, we need to convert binary
	# values to decimal
	power_val = [1, 2, 4, 8, 16, 32, 64, 128]

	val = 0
	
	for i in range(len(val_ar)):
		val += val_ar[i] * power_val[i]
		#print(val)
	return val


# In[8]:


def sp(x_train):
   
    
    for i in range(0, height):
        for j in range(0, width):
            img_lbp[i, j] = lbp_calculated_pixel(x_train, i, j)
#imglbp.append(img_lbp)
    return img_lbp
    


# In[9]:


imglbps=[]
imglbp=[]
hist1=[]
hist11=[]
_,height, width= x_train.shape
img_lbp = np.zeros((height, width),np.uint8)
#img_lbp = np.zeros((height, width))
#img_lbp=(height,width)

for z in range(0,len(x_train)):
    img_lbp=sp(x_train[z])
    #imglbp=np.append(imglbp,img_lbp)
    imglbps.append(img_lbp)
    imglbp=np.array(imglbps)
    #imglbp=np.append(imglbp,img_lbp)
    #imglbp=img_lbp
    histr = cv2.calcHist([img_lbp],[0],None,[256],[0,256])
    hist11.append(histr)
    hist1=np.array(hist11)
#imglbp=img_lbp
#x = np.append(x, 25)
print("LBP Program is finished")
        #print(imglbp)
        #print(img_lbp)
        
        
#print(img_lbp7.shape)


print(imglbp.ndim)
#print(img_lbp.ndim)
#print(x_train)
print(imglbp.shape)
print(len(imglbp))
#print(len(img_lbp))
#print(img_gray)

#histr = cv2.calcHist([img_lbp],[0],None,[256],[0,256])
print(histr)
print(hist1)
print(histr.shape)
print(hist1.shape)
        #print(len(imglbp))


# In[10]:


imglbp7=[]
imglbps7=[]
hist7=[]
hist117=[]
_,height, width= x_test.shape
img_lbp7 = np.zeros((height, width),np.uint8)
#img_lbp = np.zeros((height, width))
#img_lbp=(height,width)

for y in range(0,len(x_test)):
    img_lbp7=sp(x_test[y])
    #imglbp=np.append(imglbp,img_lbp)
    imglbps7.append(img_lbp7)
    #imglbp7=np.array(imglbps7)
    imglbp7=np.array(imglbps7)
    histr17 = cv2.calcHist([img_lbp7],[0],None,[256],[0,256])
    hist117.append(histr17)
    hist7=np.array(hist117)
    #hist7.append(histr17)
#imglbp=img_lbp
#x = np.append(x, 25)
print("LBP Program is finished")
                #print(imglbp7)
                #print(img_lbp7)
#print(x_train)

        
        
print(img_lbp7.shape)
print(img_lbp7)
#print(img_gray)

#histr = cv2.calcHist([img_lbp],[0],None,[256],[0,256])
#print(histr17)
#print(hist7)
#print(histr17.shape)
#print(len(hist7))
#print(len(imglbp7))

print(histr)
print(hist7)
print(histr.shape)
print(hist7.shape)


# In[11]:


hist1 = hist1.reshape(hist1.shape[0], hist1.shape[1]*hist1.shape[2])
hist7 = hist7.reshape(hist7.shape[0], hist7.shape[1]*hist7.shape[2])
#print(y_train.shape[0])
y_train = y_train.reshape( y_train.shape[0]*y_train.shape[1])
y_test = y_test.reshape( y_test.shape[0]*y_test.shape[1])

print("x_train: ",hist1.shape)
print("x_test: ",hist7.shape)
print("y_train: ",y_train.shape)
print("y_test: ",y_test.shape)

#print(ndim(x_train))


# In[12]:


list_names = []
list_accuracy = []


# In[13]:


#LogisticRegression#
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(hist1, y_train)
LR_accuracy = round(lr.score(hist7, y_test)*100,2)

print("LR_accuracy is %", LR_accuracy)

list_names.append("Logistic Regression")
list_accuracy.append(LR_accuracy)


# In[14]:


#Random Forest#
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(n_estimators = 400, random_state = 1)
rf.fit(hist1, y_train)
RF_accuracy = round(rf.score(hist7, y_test)*100,2)

print("RF_accuracy is %", RF_accuracy)

list_names.append("Random Forest")
list_accuracy.append(RF_accuracy)


# In[15]:


# KNeighborsClassifier#
from sklearn.neighbors import KNeighborsClassifier

Knn = KNeighborsClassifier(n_neighbors = 1) # n_neighbors=1 gives the best result for this data
Knn.fit(hist1, y_train)
Knn_accuracy = round(Knn.score(hist7, y_test)*100,2)

print("Knn_accuracy is %", Knn_accuracy)

list_names.append("KNN")
list_accuracy.append(Knn_accuracy)


# In[16]:


from sklearn.svm import SVC
# Instantiate the Support Vector Classifier (SVC)
svc = SVC(C=1.0, random_state=1, kernel='linear')
 
# Fit the model
svc.fit(hist1, y_train)


# In[17]:


from sklearn import metrics
# Make the predictions
y_predict = svc.predict(hist7)
 
# Measure the performance
print("Accuracy score %.3f" %metrics.accuracy_score(y_test, y_predict))
```

```python
import os
from skimage import io, color
import cv2
import imghdr

def prepare_training_data(data_folder_path, new_width=64, new_height=64):
    faces = []
    labels = []

    dirs = os.listdir(data_folder_path)

    for dir_name in dirs:
        label = dir_name
        subject_dir_path = os.path.join(data_folder_path, dir_name)
        subject_images_names = os.listdir(subject_dir_path)

        for image_name in subject_images_names:
            image_path = os.path.join(subject_dir_path, image_name)

            # Check if the file is a valid image
            if imghdr.what(image_path) is not None:
                img = io.imread(image_path)
                
                # Resize the image
                img_resized = cv2.resize(img, (new_width, new_height))
                img_gray = color.rgb2gray(img_resized)

                if img_gray is not None:
                    faces.append(img_gray)
                    labels.append(label)

    cv2.destroyAllWindows()
    cv2.waitKey(1)
    cv2.destroyAllWindows()

    print(labels)

    return faces, labels
```
